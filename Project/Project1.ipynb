# Home Price Prediction model

## importing model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data={
    "price":[2000,5000,7000,10000],
    "area":[200,400,600,800]
}
df=pd.DataFrame(data)
df
```

```python
plt.xlabel('area(sq fit area)')
plt.ylabel('price(Rupees)')
plt.scatter(df.area,df.price,color='blue',marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
input_value = np.array([300]).reshape(-1, 1)
reg.predict(input_value)
```

```python
d={
    "area":[202,305,456,600,700]
}
d1=pd.DataFrame(d)
d1
```

```python
p=reg.predict(d1)
d1['price']=p
```

```python
d1
```

```python
d1.to_csv('predict.csv')
```
