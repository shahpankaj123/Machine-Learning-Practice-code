```python
import pandas as pd
```

```python
df=pd.read_csv('Data_set/sport_data.csv')
df.head()
```

```python
df['sex']=df['Sex'].apply(lambda x: 1 if x=='Male' else 0)
df
```

```python
X=df.drop(['Sex','Sport'],axis='columns')
X
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle.Sport=le.fit_transform(dfle.Sport)
dfle
```

```python
Y=dfle.Sport
Y
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X,Y)
```

```python
model.score(X,Y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=30)
```

```python
model1.fit(X,Y)
```

```python
model1.score(X,Y)
```

```python
from joblib import dump
```

```python
rp=dump(model1 ,'sport_pred_model.joblib')
```
