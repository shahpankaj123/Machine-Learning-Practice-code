# One Hot Encoding

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Data_set/data.csv')
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
final_df=merged.drop(['town','pokhara'],axis='columns')
final_df
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
```

```python
X=final_df.drop(['price'],axis='columns')
X
```

```python
y=final_df.price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,0,0]])
```

```python
model.score(X,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```
