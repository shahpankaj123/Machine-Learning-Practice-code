# Saving Model as a File

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('predict.csv')
df
```

```python
plt.xlabel('area(sq fit area)')
plt.ylabel('price(Rupees)')
plt.scatter(df.area,df.price,color='blue',marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
input_value = np.array([300]).reshape(-1, 1)
reg.predict(input_value)
```

```python
from joblib import dump, load
rp=dump(reg, 'price_pred_model.joblib')
```

```python
clf = load('price_pred_model.joblib') 
input_value = np.array([300]).reshape(-1, 1)
clf.predict(input_value)
```
